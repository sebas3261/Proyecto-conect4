```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('logs/training_results.csv')
df.head()
```

```python
# Grafico de victorias por jugador
win_counts = df['winner'].value_counts()
plt.figure()
win_counts.plot(kind='bar')
plt.title('Victorias por jugador')
plt.xlabel('Jugador')
plt.ylabel('Número de victorias')
plt.show()
```

```python
# Movimientos por partida
plt.figure()
df['moves'].plot(kind='hist', bins=20)
plt.title('Distribución de movimientos por partida')
plt.xlabel('Movimientos')
plt.show()
```
